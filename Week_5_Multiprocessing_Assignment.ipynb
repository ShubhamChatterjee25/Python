### 1 . What is multiprocessing and its uses

##### Multiprocessing :

it refers to the ability of system that supports more than one processor at same time. The operating system allocates these threads to processors improving the performance of the system.

##### Uses of Multiprocessing

- Allows breaking of application into smaller threads

- The threads can run independently 

- It allows system to run multiple processes simultaneously

- Allowing to bypass "Global interpreter lock" that comes with multithreading

- Supports concurrency in both local and remote types

### 2 . Differences between Multithreading and Multiprocessing

### 3 . Write a code using multiprocessing

```python
## importing multiprocessing module
import multiprocessing

## defining a function
def introduction():
    print( "This is an external function")
    
## implementing multiprocessing
if __name__ == "__main__":
    m1 = multiprocessing.Process(target=introduction)
    print("This is Multiprocess function")
    
    ## implementing start and join
    m1.start()
    m1.join()
```

### 4 . What is Multiprocessing Pool and why is it used ?

Multiprocessing pool can be used for parallel execution of function across multiple input variables . distributing the data across processes (also called as data parallelism ).

it means it takes multiple inputs and executes them parallely across functions at same time simultaneosly. 

##### Why is it used ?

- It is a bult in package that allows the system to run multiple processes simultaneously.

- it is usefeul for cpu bound tasks such as computaionally heavy tasks

- it will benefit from multiple processors, multicore computers work faster than single core.

- it can be used for parallel execution of function across multiple input values

- load balancing , incase of cpu bound heavy tasks the processes are distributed in a way to reduce burden and ease of execution

- Gives better performance depending on the machine. the computation can be done faster.

### 5 . Creating pool of worker processes using multiprocessing module

Creating a pool worker process to print out squares of a elements inside a list.

```python
## importing multiprocessing
import multiprocessing 

## creating a simple function
def square(n):
    return n**2

## creating instance 
if __name__ == '__main__':
    
    ## writing multiprocessing to create pool
    with multiprocessing.Pool(processes = 5) as pool:
        
        ## creating variable with input and pool map 
        output = pool.map(square,[1,5,3,675,32,9])
        
        ## printing result
        print(output)
        
    
```

### 6 . Python code to create 4 processes and each print out different numbers

```python
## importing multiprocessing module
import multiprocessing

## creating a simple function
def number(n):
    print(n)

## creating instance and numbers list
if __name__ == '__main__':
    num_list = [1,2,3,4]
    
    ## creating 4 processes
    with multiprocessing.Pool(processes=4) as pool:
        
        ## applying number function to each num_list
        pool.map(number,num_list)
        
        ## closing and waiting for processes to finish
        pool.close()
        pool.join()
```
