### 1 . Explain GET and POST methods

##### GET :
- GET method is used to retrive data from server.                                               
  for example and Ecommerce application. there is a seearch button to search for an item, after entering the keyword if we noticed the keyword you searched gets displayed in the URL. This request is used when data is not sensitive.

#####  POST :
- This method enables users to send data over to the server .                             
  for example in any authentication enabled application, the registration and login form is the best example for the post method. whenever we enter information and submit, the data gets transfered over to the POST request and if we noticed, unlike the get request there is no information displayed in the URL.

##### Syntax for GET and POST methods :

- @app.routes("/example", methods = ['GET', 'POST'])

### 2 . Why is request is used in flask


- In a Flask app, we have our own web page (client) and a server. The server should process the data. The request, in FLask, is an object that contains all data sent from client to server. this data can be recovered using GET/POST methods. 

### 3 . Why is redirect( ) used in Flask ?

- Flask has redirect( ) function, when called, it return a response object and redirects the user to another target location with specified status code. location parameter is the URL where response should be redirected. status code sent to browser header defaults to 302.  

### 4 . What are templates in Flask ? and why render_template ( )is used ?

##### Templates : 
- Templates are files that contain static data as well as place holders for dynamic data. a template is rendered with specific data to produce a final document. Flask uses jinja template library to render templates in your applications, we use this to render HTML which will display in the user's browser.

##### render_template( ) :
- render_template is a flask function from the flask templating package. render_template is used to generate ouput from a template based file based on the jinja2 engine that is found in the applications templates folder. note that render_template is typically imported directly from flask package instead of from flask.
