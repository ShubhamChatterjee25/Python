### 2 . Write python code to re-index with new index that increments by 2 each row.

```python
## importing numpy
import numpy as np

## creating columns
column = ['A','B','C']

#create DataFrame
df1 = pd.DataFrame(np.random.rand(10, 3),columns= column)

## creating index with increments of 2
df1.index = np.arange(start=1,stop=len(df1)+20, step=3)## using numpy to increment indexes

#view updated DataFrame
print(df1)
```

### 10 . Create a column that contains average of sales and size of 7 including current day.

```python
## creating dataframe
df7 = pd.DataFrame({'Sales':[1000,1500,500,200,1700,800,2000,400],'Date':['2023-03-12',
'2023-06-28','2023-09-14','2023-01-05','2023-08-02','2023-04-19','2023-11-09','2023-07-22']})
```

```python
## creating new column 
df7['MovingAverage'] = df7['Sales'].rolling(7).mean() ## using rolling() method to get MA
```

```python
## seeeing the results and removing the NaN values
df7.dropna()
```

### 11 . Write python function that creates new column that contains weekday names

```python
## creating a dataframe
df8= pd.DataFrame({'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04',
'2023-01-05']})
```

```python
## converting the string to date and time series
df8['Date'] = pd.to_datetime(df8['Date'])

## creating new column and labeling the weekdays init
df8['Weekday'] = df8['Date'].dt.day_name()
```

```python
df8
```

### 12 . Write a program to select all rows between '2023-01-01' and '2023-01-31'

```python
## creating a series with random dates
s = pd.date_range('2023-01-01','2023-02-28',freq='D')

## assigning column name
column = ['Date']

## creating a dataframe 
df9= pd.DataFrame(s,columns=column)

## selecting all rows from '2023-01-01' to '2023-01-31'
df9[(df9['Date'] > '2023-01-01') & (df9['Date'] < '2023-02-01')]
```
