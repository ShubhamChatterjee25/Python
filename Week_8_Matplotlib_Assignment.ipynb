### 2 . What is a scatter plot ? use given data to generate scatter plot.

```python
## given data
import numpy as np
np.random.seed(3)
x = 3+ np.random.normal(0,2,50)
y = 3+ np.random.normal(0,2,len(x))
```

##### Scatter plot :-
- A scatter plot is a type of plot or mathematical diagram using cartesian coordinates to display values for typically two variables for a set of data. it is a graph that present relationship between two variables in a data set. if the points are coded. one additoinal varible can be displayed. 

##### Generating Scatter plot using given data

```python
## importing matplotlib
import matplotlib.pyplot as plt
```

```python
## plotting the given data into a scatter plot using matplotlib
plt.scatter(x,y)
plt.xlabel ("This is x data") ## adding xlabel
plt.ylabel ("This is y data") ## adding ylabel
plt.title ("This is a scatter plot") ## adding title
```

### 3 . Why is subplots( ) used in matplotlib ? draw given four lines using subplot( )

```python
## given four lines for subplot are 
## line 1
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

## line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

## line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

## line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

##### Why is subplot(  ) used :-
- The matplotlib.pyplot.subplots method provides a way to plot multiple plots on single figure. Given the number of rows and colums. it returns a tuple (fig, ax), giving a single figure with an array of axes ax.

```python
## plotting lineplot with given subplot( )
## creating a figure
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x1,y1)
ax.plot(x2,y2)
ax.plot(x3,y3)
ax.plot(x4,y4)
plt.title("Plotting 4 Lines of subplot")
```

### 4 . What is a bar plot ? why is it used ? use given data to play horizontal bar plot

```python
## Given data
company = np.array(["Apple","Microsoft","Google","AMD"])
profit = np.array([3000,8000,1000,10000])
```

##### Bar plot:-
- a bar plot or bar char represents a category of data with rectangular bars with length and heigth that is propotional to value they represent. they can be plotted vertically or horizontally. its sometimes called column chart.

##### Why is it used :-
- A bar plot is used when you want to show distribution of data points to perform a comparision of metric values across different subgroup of your data. from bar plot we can see which groups are highest or most common and how other groups compare against the others.

```python
## plotting horizontal bar plot
plt.barh(company,profit) ## plt.bar is normal, plt.barh is horizontal bar plot
plt.show()
```

### 5 . What is a box plot ? why is it used ? use given data to plot a box plot

```python
## given data 
box1 = np.random.normal(100,10,200)
box2 = np.random.normal(90,20,200)
```

##### Box plot :-
- A box plot is a special type of diagram that shows the quartiles in a box and the line extending from the lowest to the highest value.

##### Why is it used :-
- box plot provide quick visual summary of the variability of values in a dataset . they show median, upper, and lower quartiles, minimum and maximum vaules, and any outliers in the dataset. outerliers can reveal mistakes or unusual occurances in data.

```python
## plotting box plot 1
plt.boxplot(box1)
plt.show()
```

```python
## plotting box plot 2 
plt.boxplot(box2)
plt.show()
```
